```python
from Crawler_module.IrCrawler import IrCrawler
from Crawler_module.CpiCrawler import CpiCrawler
from Crawler_module.BillCrawler import BillCrawler
from Crawler_module.BitcoinCrawler import BitcoinCrawler
from Crawler_module.crawler_gold import Gold_Crawler
from Crawler_module.TrendsCrawler import TrendsCrawler
from Crawler_module.SPXCrawler import SPXCrawler

import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
df_ir = IrCrawler().crawl()
df_cpi = CpiCrawler().crawl()
df_bill = BillCrawler().crawl()
df_bitcoin = BitcoinCrawler().crawl()
df_gold = Gold_Crawler().crawl()
df_spx = SPXCrawler().crawl()
```

```python

def create_Chart(**kwargs):
    dfs = kwargs['dfs']
    titles = kwargs['titles']

    fig = go.Figure()

    for idx in range(len(dfs)):
        fig.add_trace(go.Scatter(x=dfs[idx][dfs[idx].columns[0]], y=dfs[idx][dfs[idx].columns[1]], name=titles[idx], yaxis= 'y' if idx == 0 else f'y{idx+1}'))


    fig.update_layout(
        
        yaxis = dict(
            title=titles[0]
        ),
        yaxis2=dict(
            title=titles[1],
            overlaying = "y",
            side="right",
        ) if len(titles) > 1 else None,
        yaxis3=dict(title="yaxis3 title", anchor="free", overlaying="y", autoshift=True) if len(titles) > 2 else None,
        yaxis4=dict(
            title="yaxis4 title",
            anchor="free",
            overlaying="y",
            autoshift=True,
            shift=-100,
        ) if len(titles) > 3 else None,
    )

    text = ''

    for title in titles:
        text += title + ' '

    fig.update_layout(
        title_text= text + '차트'
    )

    fig.show()
```

```python
dic = {'dfs':[df_gold, df_bill],
       'titles':['금', '채권']}

create_Chart(**dic)
```
